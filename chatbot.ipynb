```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import requests
from bs4 import BeautifulSoup

nltk.download('punkt')
nltk.download('stopwords')


website_url = 'https://example.com'


def scrape_website_text(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        website_text = ' '.join([p.get_text() for p in soup.find_all('p')])
        return website_text
    except Exception as e:
        print(f"Error: {e}")
        return None


documentation_text = scrape_website_text("https://www.tech-faq.com/scada.html")

if documentation_text:

    sentences = sent_tokenize(documentation_text)
    stop_words = set(stopwords.words('english'))


    indexed_documentation = {}
    for i, sentence in enumerate(sentences):
        words = word_tokenize(sentence.lower())
        words = [word for word in words if word.isalnum() and word not in stop_words]
        for word in words:
            if word in indexed_documentation:
                indexed_documentation[word].append(i)
            else:
                indexed_documentation[word] = [i]

    def find_sentences(keyword):
        keyword = keyword.lower()
        if keyword in indexed_documentation:
            return [sentences[i] for i in indexed_documentation[keyword]]
        else:
            return []

    
    while True:
        user_input = input("Ask a question or type 'exit' to quit: ")

        if user_input.lower() == 'exit':
            print("Goodbye!")
            break

        
        sentences_containing_keyword = find_sentences(user_input)

        if sentences_containing_keyword:
            print("Here are sentences containing the keyword:")
            for sentence in sentences_containing_keyword:
                print("- " + sentence)
        else:
            print("Keyword not found in the documentation.")
else:
    print("Failed to scrape the website. Please check the URL.")
```
